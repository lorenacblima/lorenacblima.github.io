```python
#formatando os dados
type_label=np.where(recipes['Type']=='Muffin',0,1)
recipe_features=recipes.columns.values[1:].tolist()
recipe_features
ingredients=recipes[['Flour', 'Sugar']].values
print(ingredients)
```

```python
#fit

model=svm.SVC(kernel='linear')
model.fit(ingredients, type_label)
```

```python
#separando o hiperplano
w=model.coef_[0]
a= -w[0]/w[1]
xx=np.linspace(30,60)
yy=a*xx-(model.intercept_[0])/w[1]
print(yy)
#Plotando as paralelas que separam o hiperplano
b=model.support_vectors_[0]
yy_down=a*xx+(b[1]-a*b[0])
b=model.support_vectors_[1]
yy_up=a*xx+(b[1]-a*b[0])
print(yy_down)
print(yy_up)
```

```python
sns.lmplot('Flour', 'Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={'s':70});
plt.plot(xx,yy,linewidth=3, color='orange')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
```

```python
sns.lmplot('Egg', 'Milk', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={'s':70});
plt.plot(xx,yy,linewidth=3, color='orange')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
```

```python
sns.lmplot('Milk', 'Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={'s':70});
plt.plot(xx,yy,linewidth=3, color='orange')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
```

```python
sns.lmplot('Flour', 'Flour', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={'s':70});
plt.plot(xx,yy,linewidth=3, color='orange')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
```

```python
#criando uma função para predizer se é cupcake ou muffin


def muffin_or_cupcake(flour,sugar):
    if(model.predict([[flour, sugar]]))==0:
        print('Isso é uma receita de muffin.')
    else:
        print('Isso é uma receita de cupcake.')
        
muffin_or_cupcake(70,30)

#plot
sns.lmplot('Flour', 'Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={'s':70});
plt.plot(xx,yy,linewidth=3, color='orange')
plt.plot(70, 30, 'yo', markersize=10)
```
